```python
import pandas as pd
```

```python
df = pd.read_csv("vine.csv", names=["review_id", "star_rating", "helpful_votes", "total_votes","vine","verified_purchase"])
df.head()
```

```python
# Filter the data and create a new DataFrame or table to retrieve all the rows where the total_votes count 
# is equal to or greater than 20 to pick reviews that are more likely to be helpful and to avoid having 
# division by zero errors later on.

total_votes_filter_df = df[df['total_votes']>20]
total_votes_filter_df.head()
```

```python
# Filter the new DataFrame or table created in Step 1 and create a new DataFrame or table to retrieve all the rows 
# where the number of helpful_votes divided by total_votes is equal to or greater than 50%.

helpful_votes_filter_df = total_votes_filter_df[total_votes_filter_df['helpful_votes']/df['total_votes']>=.5]
helpful_votes_filter_df.head(20)
```

```python
# Filter the DataFrame or table created in Step 2, and create a new DataFrame or table that retrieves all the rows where 
# a review was written as part of the Vine program (paid), vine == 'Y'.

in_vine_df = helpful_votes_filter_df[helpful_votes_filter_df['vine']=='Y']
in_vine_df.head()
```

```python
not_in_vine_df = helpful_votes_filter_df[helpful_votes_filter_df['vine']=='N']
not_in_vine_df.head()
```
